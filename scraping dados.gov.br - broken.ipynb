```python
import urllib.request
import json
import pprint
import pandas as pd
```

```python
# Make the HTTP request.
response = urllib.request.urlopen('http://dados.gov.br/api/action/package_list')
assert response.code == 200

# Use the json module to load CKAN's response into a dictionary.
response_dict = json.loads(response.read())

# Check the contents of the response.
assert response_dict['success'] is True
result = response_dict['result']
pprint.pprint(result)
```

```python
ct=0
json_packages = []
for id_dataset in result:
    print(id_dataset)
    response = urllib.request.urlopen('http://dados.gov.br/api/action/package_show?id=' + id_dataset)
    response_dict = json.loads(response.read())
    result_package = response_dict['result']
    json_packages.append(result_package)
    #if(ct==2):
    #    break
    #ct+=1
```

```python
#pprint.pprint(json_packages[0])
```

```python
df = pd.concat([pd.DataFrame(x['resources']) for x in json_packages], ignore_index=True)
```

```python
df.to_csv('dadosgovbr-recursos.csv')
```

```python
with open('dump-json.txt', 'w') as f:
    for item in json_packages:
        f.write("%s\n" % item)
```
